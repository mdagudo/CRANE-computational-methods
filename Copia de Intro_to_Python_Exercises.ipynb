## **Week 1: Syntax, Variables & Functions**

#### Exercise 1

A. Define 5 variables with the values -11.3, 24, 1, 3.4, and -5 as floats. Confirm they are floats by printing their varibale type.

B. Define a function that does the following:

1.   Takes 3 numbers and calculates their sum.
2.   Convert that sum to an int and check if the value of this int is greater than or less than 0 (Use if/else statements).
3.   Return the string "Greater/Less than zero" depending on the value of the int.

C. Document your function using docstring format of your choice. Make sure to define the variable types in your documentation.

D. Print the docstring for your function

E. Test your function by printing out what your function returns when different combinations of the variables you defined in part A are passed in as arguments.


```python
# your solution here
a = -11.3
b = float(24)
c = float(1)
d = 3.4
e = float(-5)

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))

def sum(x, y, z):
  """
  Sums three given numbers
  and checks if it is greater or less than zero.

Parameters
------------
  x: float
  first number
  y: float
  second number
  z: float
  third number

Returns
------------
  w: integer
  sum of x, y and z
  """
  w = int(x + y + z)

  if w > 0:
    print("Greater than zero")
  else:
    print("Less than zero")

  return w

print(sum.__doc__)

print(sum(a, b, c))
print(sum(a, b, d))
print(sum(a, b, e))

```

#### Exercise 1

   A. Create an array containing the values 4, 0, 6, 5, 11, 14, 12, 14, 5, 16.
   
   B. Create a 10x2 array of zeros and name it zerosarray.
   
   C. Write a for loop that checks if each of the numbers in the first array squared is less than 100. If the statement is true, change that row of your zeros array to equal the number and its square. Hint: you can change the value of an array by stating "zerosarray[i] = [a number, a number squared]".
   
   D. Print out the final version of your zeros array.
    
Hint: should you loop over the elements of the array or the indices of the array?

```python
array = np.array([4, 0, 6, 5, 11, 14, 12, 14, 5, 16])
zerosarray = np.zeros((10,2))

print('Zeros array: \n', zerosarray)

for i in range(len(array)):
  if array[i]**2 < 100:
    zerosarray[i] = [array[i], (array[i])**2]

print('Final version of array: \n', zerosarray)
```

Exercise 1:

1. Create 2 arrays, both which go from 1 to 50.
2. Multiply the first array by 100 and add 50. Multiply the second by 50 and add 75 to it.
3. Create a 2x2 grid of subplots.

  A. In the top left plot the two arrays.

  B. In the top right subtract the two arrays.

  C. In the bottom left multiply the two arrays.

  D. In the bottom right make a scatter plot of the multiplication of the two arrays

  E. Make sure that each plot has a different color, line thickness, line style, and that all the plots have a label on the X and Y axis and a legend for the first plot.


```python
# your solutions here
x = np.linspace(1, 50, 50)
y = np.linspace(1, 50, 50)

z = x*100 + 50
w = y*50 + 75

plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.plot(z, linestyle="dashed", linewidth="1", color="green")
plt.plot(w, linestyle="dotted", linewidth="2", color="yellow")
plt.legend(['z', 'w'])
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,2)
plt.plot(z-w, linewidth="3", color="red")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,3)
plt.plot(z*w, linewidth="4")
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2,2,4)
plt.scatter(z, w, color="pink")
plt.xlabel('x')
plt.ylabel('y')
```

Exercise 2:

1. Create 2 arrays, one from 0 to 2*pi with 50 entries (call it x), the other from 0 to 4*pi with 100 entries (call it y).
2. Use the meshgrid function to create a meshgrid of these two functions.
3. Make a third function which is sin(meshgrid(x)) + cos(meshgrid(y))
4. Plot this function.
5. Create contours on your plot.
6. Create a label for the x axis, y axis, and title.
7. Use the viridis colormap.

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.linspace(0, 4*np.pi, 100)

xx, yy = np.meshgrid(x, y) #Generates the x- and y-coordinate arrays

z = np.sin(xx) + np.cos(yy)

plt.contourf(z, cmap="viridis")              #Plots the data to show the 2D image
plt.colorbar()             #This gives us the value for each color plotted
plt.xlabel("x")
plt.ylabel("y")
plt.title("Contour Plot")

plt.show()
```
